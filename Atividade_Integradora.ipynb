```python
import pandas as pd

file_path = 'Dados_tel.csv'

data_frame = pd.read_csv(file_path)

df = pd.read_csv(file_path)

print(df)

def avaliar_lancamento(df_data):
    df_data['pressao_tanque_bar'] = pd.to_numeric(df_data['pressao_tanque_bar'], errors='coerce')

    estado = df_data.iloc[-1]

    motivos = []

    if estado["integridade_estrutural"] == 0:
        motivos.append("Falha estrutural")

    if estado["temperatura_interna_c"] > 35:
        motivos.append("Temperatura interna crítica")

    if estado["nivel_energia_pct"] < 60:
        motivos.append("Energia insuficiente")

    if estado["pressao_tanque_bar"] < 4.0:
        motivos.append("Pressão dos tanques baixa")

    modulos = [
        "modulo_vida",
        "modulo_propulsao",
        "modulo_comunicacao",
        "modulo_termico"
    ]

    for m in modulos:
        if estado[m] == "FAIL":
            motivos.append(f"Falha no {m}")

    # Resultado final
    if len(motivos) == 0:
        return "✅ LANÇAMENTO PERMITIDO", []
    else:
        return "❌ LANÇAMENTO NÃO PERMITIDO", motivos


# Exemplo de uso
status, problemas = avaliar_lancamento(df)

print(status)

if problemas:
    print("\nMotivos:")
    for p in problemas:
        print("- ", p)
```

## 1. Classificação dos dados

###  Variáveis contínuas (numéricas)

*   Temperatura interna (°C)
*   Temperatura externa (°C)
*   Nível de energia (%)
*   Pressão do tanque (bar)

**Características:**

*   Séries temporais
*   Tendência clara (não estacionárias)
*   Forte correlação entre variáveis (energia, pressão e temperatura)

###  Variáveis binárias

*   Integridade estrutural (0/1)

**Características:**

*   Indicador de estado crítico
*   Mudança abrupta (evento de falha)

###  Variáveis categóricas (estado)

**Módulos:**

*   Vida
*   Propulsão
*   Comunicação
*   Térmico

**Valores:**

*   OK
*   WARN
*   FAIL

**Características:**

*   Sistema de estados discretos
*   Evolução sequencial (OK → WARN → FAIL)

---

## 2. Identificação de possíveis anomalias

### Anomalia térmica

*   Temperatura interna sobe continuamente (22 → 48°C)
*   Enquanto externa cai (-55 → -85°C)

**Problema:**

*   Comportamento inverso ao esperado → falha de controle térmico

### Anomalia energética

*   Queda acelerada e constante (25%/h)

**Problema:**

*   Consumo excessivo não justificado apenas por operação normal
*   Provável sobrecarga de sistemas tentando compensar falhas

### Anomalia de pressão

*   Queda contínua (5.2 → 2.9 bar)

**Problema:**

*   Indício de:
    *   vazamento
    *   falha de vedação
    *   consumo anormal

### Anomalia nos módulos (falha em cascata)

**Sequência observada:**

*   Térmico → WARN → FAIL
*   Comunicação → degrada
*   Propulsão → falha
*   Vida → falha final

**Problema:**

*   Forte dependência entre sistemas
*   Falha inicial propagando para todo o sistema

### Anomalia estrutural

*   Integridade muda de 1 → 0 abruptamente

**Problema:**

*   Indica evento crítico (colapso)
*   Provavelmente consequência acumulada:
    *   pressão baixa
    *   estresse térmico

---

## 3. Sugestões de risco

### Risco alto (crítico)

**1. Falha do sistema térmico**

*   **Impacto:** global
*   **Consequência:**
    *   superaquecimento interno
    *   degradação de equipamentos

**Ação sugerida:**

*   Redundância térmica
*   Monitoramento preditivo com alerta precoce

**2. Perda de pressão**

*   **Impacto:** estrutural + operacional

**Ação sugerida:**

*   Sensores redundantes
*   Sistema automático de isolamento de vazamento

**3. Colapso energético**

*   **Impacto:** todos os sistemas

**Ação sugerida:**

*   Controle de carga inteligente (load shedding)
*   Priorização de módulos críticos (vida > comunicação > outros)

### Risco médio

**4. Falha em cascata de módulos**

*   **Problema:** sistêmico (dependência excessiva)

**Ação sugerida:**

*   Arquitetura desacoplada
*   Fail-safe independente por módulo

**5. Falha estrutural tardia**

*   Ocorre após degradação prolongada

**Ação sugerida:**

*   Modelos preditivos baseados em:
    *   temperatura
    *   pressão
    *   vibração (se disponível)

### Risco operacional

**6. Falta de resposta adaptativa**

*   Sistema não reage às falhas iniciais

**Ação sugerida:**

*   Implementar:
    *   controle autônomo
    *   IA de detecção de anomalias em tempo real
