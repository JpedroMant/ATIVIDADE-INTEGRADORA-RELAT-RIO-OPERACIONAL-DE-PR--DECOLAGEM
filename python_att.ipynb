```python
import pandas as pd

file_path = 'Dados_tel.csv'

data_frame = pd.read_csv(file_path)

df = pd.read_csv(file_path)

print(df)

def avaliar_lancamento(df_data):
    df_data['pressao_tanque_bar'] = pd.to_numeric(df_data['pressao_tanque_bar'], errors='coerce')

    estado = df_data.iloc[-1]

    motivos = []

    if estado["integridade_estrutural"] == 0:
        motivos.append("Falha estrutural")

    if estado["temperatura_interna_c"] > 35:
        motivos.append("Temperatura interna crítica")

    if estado["nivel_energia_pct"] < 60:
        motivos.append("Energia insuficiente")

    if estado["pressao_tanque_bar"] < 4.0:
        motivos.append("Pressão dos tanques baixa")

    modulos = [
        "modulo_vida",
        "modulo_propulsao",
        "modulo_comunicacao",
        "modulo_termico"
    ]

    for m in modulos:
        if estado[m] == "FAIL":
            motivos.append(f"Falha no {m}")

    # Resultado final
    if len(motivos) == 0:
        return "✅ LANÇAMENTO PERMITIDO", []
    else:
        return "❌ LANÇAMENTO NÃO PERMITIDO", motivos


# Exemplo de uso
status, problemas = avaliar_lancamento(df)

print(status)

if problemas:
    print("\nMotivos:")
    for p in problemas:
        print("- ", p)
```
